```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Supongamos que df es tu DataFrame con las columnas 'Años' y 'PorcentajeTriples'
# Asegúrate de tener un índice temporal para trabajar con series temporales
tabla_resultado_P['Year'] = pd.to_datetime(tabla_resultado_P.index, format='%Y')
tabla_resultado_P.set_index('Year', inplace=True)

# Normalizar los datos
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(tabla_resultado_P[['3P%']])

# Crear conjuntos de entrenamiento y prueba
train_data = scaled_data[:-5]  # Tomar todos menos los últimos 5 años
test_data = scaled_data[-5:]   # Tomar los últimos 5 años para la evaluación

# Función para preparar los datos en secuencias para LSTM
def create_sequences(data, seq_length):
    sequences = []
    targets = []
    for i in range(len(data) - seq_length):
        seq = data[i:i + seq_length]
        label = data[i + seq_length]
        sequences.append(seq)
        targets.append(label)
    return np.array(sequences), np.array(targets)

# Definir la longitud de la secuencia
sequence_length = 3  # Ajusta según sea necesario

# Crear secuencias para entrenamiento y prueba
X_train, y_train = create_sequences(train_data, sequence_length)
X_test, y_test = create_sequences(test_data, sequence_length)

# Crear el modelo LSTM
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(sequence_length, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Reshape los datos para que coincidan con la entrada de LSTM [samples, time steps, features]
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Entrenar el modelo
model.fit(X_train, y_train, epochs=100, verbose=1)

# Extender el conjunto de datos para los próximos 5 años
future_years = pd.date_range(tabla_resultado_P.index[-1] + pd.DateOffset(years=1), periods=5, freq='Y')

# Normalizar los datos de prueba extendidos
scaled_future_data = scaler.transform(tabla_resultado_P.loc[future_years][['3P%']])
X_future, y_future = create_sequences(scaled_future_data, sequence_length)

# Reshape los datos para que coincidan con la entrada de LSTM [samples, time steps, features]
X_future = X_future.reshape((X_future.shape[0], X_future.shape[1], 1))

# Realizar predicciones para los próximos 5 años
future_predictions = model.predict(X_future)

# Desnormalizar las predicciones
future_predictions = scaler.inverse_transform(future_predictions)

# Crear un rango de años para las predicciones futuras
years_future = future_years.year

# Graficar resultados
plt.plot(tabla_resultado_P.index[-5:], tabla_resultado_P['3P%'][-5:], label='Datos de prueba', color='orange')
plt.plot(years_future, future_predictions, label='Predicciones futuras', color='green')
plt.xlabel('Year')
plt.ylabel('3P%')
plt.legend()
plt.show()
```
