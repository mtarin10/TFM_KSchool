### Regresión polinómica de grado superior y SVR

```python
# Crea un nuevo DataFrame con el nombre de la columna y los valores de la media
tabla_resultado_P = pd.DataFrame({
    'Year': tabla_media_triples_P.index,
    '3P%': tabla_media_triples_P.values
})
print(tabla_resultado_P)
tabla_resultado_P.index
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split


# Supongamos que df es tu DataFrame con las columnas 'Años' y 'PorcentajeTriples'

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(tabla_resultado_P['Year'].values.reshape(-1, 1), tabla_resultado_P['3P%'].values, test_size=0.2, random_state=42)

# Regresión polinómica de grado superior
poly_degree = 3  # Ajusta según sea necesario
poly_features = PolynomialFeatures(degree=poly_degree)
X_poly_train = poly_features.fit_transform(X_train)
X_poly_test = poly_features.transform(X_test)

model_poly = LinearRegression()
model_poly.fit(X_poly_train, y_train)
y_pred_poly = model_poly.predict(X_poly_test)

# SVR
model_svr = SVR(kernel='rbf')
model_svr.fit(X_train, y_train)
y_pred_svr = model_svr.predict(X_test)

# Graficar resultados
plt.scatter(tabla_resultado_P['Year'], tabla_resultado_P['3P%'], label='Datos históricos')
plt.plot(X_test, y_pred_poly, color='red', label=f'Regresión Polinómica (Grado {poly_degree})')
plt.plot(X_test, y_pred_svr, color='blue', label='SVR')
plt.xlabel('Year')
plt.ylabel('3P%')
plt.legend()
plt.show()
```

```python
# Supongamos que df es tu DataFrame con las columnas 'Años' y 'PorcentajeTriples'

# Extender el conjunto de datos hasta los próximos 5 años
future_years = np.arange(2018, 2023).reshape(-1, 1)

# Regresión polinómica de grado superior
future_poly_features = poly_features.transform(future_years)
future_predictions_poly = model_poly.predict(future_poly_features)

# SVR
future_predictions_svr = model_svr.predict(future_years)

# Graficar resultados
plt.scatter(tabla_resultado_P['Year'], tabla_resultado_P['3P%'], label='Datos históricos')
plt.plot(X_test, y_pred_poly, color='red', label=f'Regresión Polinómica (Grado {poly_degree})')
plt.plot(X_test, y_pred_svr, color='blue', label='SVR')
plt.scatter(future_years, future_predictions_poly, color='orange', label='Predicciones futuras (Polinómico)')
plt.scatter(future_years, future_predictions_svr, color='green', label='Predicciones futuras (SVR)')
plt.xlabel('Year')
plt.ylabel('3P%')
plt.legend()
plt.show()
```
